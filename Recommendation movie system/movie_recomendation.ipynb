```python
import numpy as np
import pandas as pd
import difflib #this library use to get the closest match 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
dataset=pd.read_csv("movies.csv")
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset.shape
```

```python
selected_feature=["genres","keywords","tagline","cast","crew","original_language","overview"]
```

```python
selected_feature
```

```python
# replacing the null value from null string

for feature in selected_feature:
    dataset[feature]=dataset[feature].fillna(" ")
```

```python
selected_columns=dataset["genres"]+" "+dataset["keywords"]+" " +dataset["tagline"]+" "+dataset["cast"]+" "+dataset["crew"]+" "+dataset["original_language"]
```

```python
selected_columns
```

```python
tfidf = TfidfVectorizer()
```

```python
feature_vector=tfidf.fit_transform(selected_columns)
```

```python
print(feature_vector)
```

```python
#getting the similarity score or similarity confidence value, using cosine simalirty funciton
```

```python
simialrity=cosine_similarity(feature_vector)
```

```python
print(simialrity)
```

```python
simialrity.shape
```

```python
#getting movie name from user
```

```python
movie_name=input("enter your fav movie name")
movie_name
```

```python
#creating a alist with all the movie name 
```

```python
list_of_all_movies=dataset["title"].tolist()
```

```python
list_of_all_movies
```

```python
# finding the close max for the movie name given by the user
```

```python
find_close_max=difflib.get_close_matches(movie_name, list_of_all_movies)#difflib is use to find the close matches 

for movie in find_close_max:
    print(movie)
```

```python
close_match=find_close_max[0]
print(close_match)
```

```python
# finding the index of the movie with title
```

```python
index_of_movie = dataset[dataset.title == close_match]["index"].values[0]
index_of_movie
```

```python
index_of_movie = dataset[dataset.title == close_match].iloc[0]["index"]

index_of_movie
```

```python
#getting the list of same movie 
simialrity_score=list(enumerate(simialrity[index_of_movie]))
```

```python
simialrity_score
```

```python
#sorting the movies based on there similarity score
```

```python
sorted_similar_movier=sorted(simialrity_score, key=lambda x:x[1], reverse=True)#reverse=true is because we want it into decending order, x:x[1] means that score, if we do x:x[0] this means the index, key=lambda is to finiding the value
sorted_similar_movier
```

```python
#suggesting the name of similar movies based on the index
```

```python
print("Movies suggested for you: \n")

i = 1

for move in sorted_similar_movier:  # Assuming this contains tuples like (index, similarity_score)
    index = move[0]
    title_from_index = dataset[dataset.index == index]["title"].values[0]

    if i < 21:  # Fixing the indentation
        print(i, ":", title_from_index)
        i += 1
```

```python
#movie recommendation system
movie_name1 = input(' Enter your favourite movie name : ')

list_of_all_titles1 = dataset['title'].tolist()

find_close_match1 = difflib.get_close_matches(movie_name1, list_of_all_titles1)

close_match1 = find_close_match1[0]

index_of_the_movie1 = dataset[dataset.title == close_match1]['index'].values[0]

simialrity_score1=list(enumerate(simialrity[index_of_the_movie1]))

sorted_similar_movies1 = sorted(simialrity_score1, key = lambda x:x[1], reverse = True) 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies1:
  index1 = movie[0]
  title_from_index1 = dataset[dataset.index==index1]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index1)
    i+=1
```
